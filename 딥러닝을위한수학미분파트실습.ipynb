## 기울기 시각화

```python
import numpy as np
x = np.linspace(-1.0, 1.0, 50)
y = np.linspace(-1.0, 1.0, 50)
xx = []; yy = []; zz=[]
for i in range(len(x)):
    for j in range(len(y)):
        xx.append(x[i])
        yy.append(y[j])
        zz.append(x[i]*x[i] + x[i]*y[j]+y[j]*y[j])
x = np.array(xx)
y = np.array(yy)
z = np.array(zz)
```

```python
from mpl_toolkits.mplot3d import Axes3D # matplotlib 3차원 확장 모듈
import matplotlib.pylab as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, marker = '.', s=2, color='b')
ax.view_init(30,20)#3D 도형의 형태를 잘 볼 수 있도록 그래프 회전
plt.draw(
)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
x = np.linspace(-1.0, 1.0, 20)
y = np.linspace(-1.0, 1.0, 20)
xv, yv = np.meshgrid(x, y, indexing= 'ij', sparse=False)
dx = 2*xv + yv
dy = 2*yv + xv
ax.quiver(xv,yv,dx,dy,color='b')#화살표로 벡터장 그래프 생성
plt.axis('equal')
plt.show()
```
